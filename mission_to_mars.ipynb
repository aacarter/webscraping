```python
from bs4 import BeautifulSoup
import requests
from splinter import Browser
import pandas as pd
```

```python
url = 'https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'
```

```python
response = requests.get(url)
mars_dict={}
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
news = soup.find_all('div', class_="content_title")[0]
news_title = news.text.strip()
news_title
```

```python
news_paragraph = soup.find_all('div', class_="rollover_description_inner")[0]
news_p = news_paragraph.text.strip()
news_p
```

```python
mars_dict['news title']=news_title
mars_dict['news paragraph']=news_p
mars_dict
```

```python
image_link = 'https://www.jpl.nasa.gov'
url2 = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(url2)
```

```python
for result in results:  
    img = result.find('a')
img_url = img['href']
featured_image_url = image_link + img_url
```

```python
mars_dict['featured image url'] = featured_image_url
mars_dict
```

```python
!which chromedriver
executable_path = {'executable_path': 'chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
url5='https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
homepage = 'https://astrogeology.usgs.gov'
browser.visit(url5)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
astropedia = soup.find_all('div',class_='description')
astropedia
```

```python
astro_list = []
for a in astropedia:
    title = homepage+a.find('a')['href']
    astro_list.append(title)
astro_list
```

```python
mars_list = []
for astro in astro_list:
    browser.visit(astro)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    astro_dict = {}
    image = soup.find('img', class_ = 'wide-image')
    title = soup.find('h2')
    img = image['src']
    title_url = title.text
    img_url = homepage+img
    astro_dict['image_url'] = img_url
    astro_dict['title'] = title_url
    mars_list.append(astro_dict)
```

```python
mars_dict['mars list'] = mars_list
mars_dict
```
